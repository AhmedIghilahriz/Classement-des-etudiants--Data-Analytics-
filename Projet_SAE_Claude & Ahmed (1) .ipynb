 <div style="background-color: #f2f2f2; padding: 25px;">
  <p>Nous avons commencé par utiliser le fichier Excel reçu et, à l'aide de Python et Pandas, nous avons éliminé les candidats qui n'avaient pas validé leur candidature. Cela a réduit le nombre de candidats à traiter de <strong>4605 candidats</strong> à <strong>4362 candidats</strong>.</p>
    <p>Ensuite, nous avons séparé les candidats en <strong>deux</strong> groupes : ceux ayant obtenu un baccalauréat <strong>général</strong> et ceux ayant obtenu un baccalauréat <strong>technologique</strong>.</p>
</div>

```python
import pandas as pd 
df= pd.read_excel('PS_INFO.xlsx')
#df
```

```python
nb_candidature = len(df)
nb_non_conf=0
nb_BAC_G=0
nb_BAC_T=0

for i in range(nb_candidature) :
    groupe = df['Groupe'][i]
    if(df['Candidature validée (O/N)'][i]=='Non') : nb_non_conf+=1
    else : 
        if(groupe==7415) : nb_BAC_G+=1
        else : nb_BAC_T+=1
print("\t Nombre de voeux total :", nb_candidature)
print("\t Nombre de voeux non confirmés :", nb_non_conf)
print("\t Nombre de voeux confirmés :", nb_candidature-nb_non_conf)
print('\t Répartition :')
print("\t ", nb_BAC_G, "bacs généraux")
print("\t ", nb_BAC_T, "bacs technologique")
```

<p>Nombre de voeux total : <strong>4605 candidats</strong> <br>
    Nombre de voeux non confirmés : <strong>243 candidats</strong><br>
    Nombre de voeux confirmés : <strong>4362 candidats</strong></p>
<p>&emsp;Répartition :</p>
<p>&emsp;&emsp;<strong>2972 bacs généraux</strong></p>
<p>&emsp;&emsp;<strong>1390 bacs technologiques</strong></p>

```python
import matplotlib.pyplot as plt
def circulaire(dataframe):
    plt.figure(figsize=(20,15))
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    plt.subplot(2,3,1)
    dataframe['Sexe'].value_counts().plot.pie(autopct='%.2f')
    plt.title('Genre')
    plt.ylabel('')
    plt.subplot(2,3,2)
    dataframe['Boursier'].value_counts().plot.pie(autopct='%.2f')
    plt.title('Boursier')
    plt.ylabel('')
    plt.subplot(2,3,3)
    dataframe['Département établissement'].value_counts().plot.pie(autopct='%.2f')
    plt.title('Département')
    plt.ylabel('')
    #dataframe['Libellé établissement'].value_counts().plot.bar()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def histogramme(serie, n):
    # Créer une nouvelle colonne contenant le numéro du département et le nom de l'établissement
    serie['Nom etablissement'] = '('+ serie['Libellé établissement'] + ')'  + " " +  serie['Département établissement'].astype(str)

    # Sélectionner les n premiers établissements
    lycees = serie['Nom etablissement'].value_counts().head(n)

    # Créer le graphique
    fig = plt.figure(figsize=(30, 15))
    plt.xticks(rotation=90, fontsize=20)

    # Tracer l'histogramme
    bars = plt.bar(lycees.index, lycees.values)

    # Ajouter le titre et les étiquettes des axes
    plt.title('Nombre de candidats par établissement', fontsize=30)
    plt.xlabel('Ecoles', fontsize=25)
    plt.ylabel('Nombre de candidats', fontsize=25)
    plt.tick_params(axis='y', width=5, labelsize=20)

    # Ajouter le nombre de candidats au-dessus de chaque barre
    for i, bar in enumerate(bars):
        plt.text(bar.get_x() + bar.get_width()/2.0, bar.get_height(), 
                 str(int(bar.get_height())), ha='center', va='bottom', fontsize=15)

    # Afficher le graphique
    plt.show()
```

```python
import pandas as pd
def classer(serie, matieres, trimestres, nv_ecart_type, nv_moyenne):
    # Définition des coefficients pour chaque matière
    coef_math = [1, 2, 3, 2, 3, 4]
    coef_français = [1, 1, 1, 1, 1, 1]
    coef_anglais = [1, 1, 1, 1, 1, 1]
    # Création d'un dictionnaire de coefficients pour chaque matière
    coefs = {
        'mathématiques': coef_math,
        'français': coef_français,
        'langue vivante a' : coef_anglais
    }
    serie['les_non_classe']= False
    
    coef_globale_matiere = {'mathématiques':13,'français': 9,'langue vivante a': 4}
    # Parcours de toutes les combinaisons matière/trimestre
    for matiere in matieres:
        for trimestre in trimestres:
            # Calcul des notes normalisées pour la matière et le trimestre actuels
            x_norm = (serie[f'Moyenne candidat en {matiere} Trimestre {trimestre}'] - serie[f'Moyenne classe en {matiere} Trimestre {trimestre}']) / (serie[f'Moyenne plus haute en {matiere} Trimestre {trimestre}'] - serie[f'Moyenne plus basse en {matiere} Trimestre {trimestre}'])
            x_min = x_norm.min()
            x_max = x_norm.max()
            serie[f'normalized_note_{matiere}_T{trimestre}'] = 20 / (x_max - x_min) * (x_norm - x_min)

            # Calcul de la moyenne et de l'écart-type pour la matière et le trimestre actuels
            moyenne = serie[f'normalized_note_{matiere}_T{trimestre}'].mean()
            ecart_type = serie[f'normalized_note_{matiere}_T{trimestre}'].std()

            # Calcul des nouvelles notes normalisées pour la matière et le trimestre actuels
            serie[f'nouvelle_note_normalisee_{matiere}_T{trimestre}'] = (serie[f'normalized_note_{matiere}_T{trimestre}'] - moyenne) / ecart_type * nv_ecart_type + nv_moyenne
    
    # Calcul des notes globales pour chaque élève
    for index, row in serie.iterrows():
        #intialise la note gloabale
        note_globale=0
        
        for matiere in matieres:
            note_globale_matiere= 0
            coef_matiere = coefs[matiere.lower()]
            for j, trim in enumerate(trimestres):
               #affecter la variable "coef_trimestre" les coefitions de la matiere actuelle 
                coef_trimestre = coef_matiere[j]
                note = serie.loc[index, f'nouvelle_note_normalisee_{matiere}_T{trim}']
                #note_global egale a la note normalisé de la matiere multiplier par le coef de trimestre
                note_globale_matiere += (coef_trimestre * note)
         
            #ici on vas placer les planches si l'eleves va etre en desous de la planche, l'eleve vas etre non classer 
            #C-a-d : 'les_note_classe'=True 
            if(matiere=="Mathématiques"):
                
                note_globale_matiere = note_globale_matiere/sum(coefs['mathématiques'])
                serie.at[index,'note_mathematiques']=note_globale_matiere
                if (note_globale_matiere <10  or pd.isna(note_globale_matiere)):
                    serie.at[index, 'les_non_classe'] = True
                #la note globale avec coefficion de la matiere    
                note_globale_avc_coef= note_globale_matiere *  coef_globale_matiere['mathématiques']    
            if(matiere=="Français"):
                note_globale_matiere = note_globale_matiere/sum(coefs['français'])
                #cette variable est faite pour calculer la coreelation 
                serie.at[index,'note_francais']=note_globale_matiere
                if (note_globale_matiere < 8 or pd.isna(note_globale_matiere)):
                    serie.at[index, 'les_non_classe'] = True
                note_globale_avc_coef=note_globale_matiere * coef_globale_matiere['français']   
            if(matiere=="Langue vivante A"):
                note_globale_matiere = note_globale_matiere/sum(coefs['langue vivante a'])
                #cette variable est faite pour calculer la coreelation 
                serie.at[index,'note_anglais']=note_globale_matiere
                if (note_globale_matiere <8 or pd.isna(note_globale_matiere)):
                    serie.at[index, 'les_non_classe'] = True
                note_globale_avc_coef=note_globale_matiere*coef_globale_matiere['langue vivante a'] 
                
            #quand on calcule la note globale avec coef  on va l'affecter a une autre variable "note_global" pour cumuler tout les moyenne de l'eleve
            note_globale+= note_globale_avc_coef        
            
            
        #la moyenne sera la note globale qu'est ete cumule avec divisé par la somme des coefition de matieres
        serie.at[index, 'note_globale'] = note_globale/sum(coef_globale_matiere.values())

        
    serie = serie.sort_values('note_globale', ascending=False)
    serie = serie.reset_index(drop=True)
    serie.index += 1
    serie['index']=serie.index

    return serie[['index','ID','Sexe', 'Boursier', 'Département établissement','Libellé établissement', 'EDS BAC Terminale', 'note_globale','note_mathematiques','note_francais','note_anglais','les_non_classe', "Niveau d'étude.5","Année scolaire","Série diplôme" ]]
```

<div style="background-color: #f2f2f2; padding: 20px;">
<p>Nous avons étudié les relations entre la <strong>note</strong> de l'élève dans une matière spécifique (<strong>Mathématiques</strong>, <strong>Français</strong>, <strong>Anglais</strong>) et sa <strong>moyenne globale</strong>, ainsi que son <strong>rang</strong> dans le classement. Nous avons constaté que pour toutes les matières, si un élève obtient une excellente note, cela a un impact significatif sur sa moyenne globale et son classement dans le rang. En d'autres termes, les élèves qui obtiennent d'excellentes notes dans ces différentes matières ont tendance à avoir une <strong>meilleure moyenne globale</strong> et à être <strong>mieux classés</strong>.</p>

<p>Cette conclusion est importante car elle met en évidence l'importance de la régularité dans la réussite scolaire. En effet, des notes élevées dans juste une seule matière ou deux ne suffisent pas à garantir une bonne moyenne globale ou un classement élevé. Les élèves doivent maintenir une bonne performance dans toutes les matières pour avoir une chance d'obtenir de bons résultats et d'être bien classés.</p>

<p>En résumé, notre analyse a montré que les élèves ayant de bonnes notes dans les matières clés ont tendance à réussir mieux que les autres. Par conséquent, il est important pour les élèves de travailler régulièrement dans toutes les matières pour améliorer leur moyenne globale et leur positionnement dans le classement.</p>

```python
import matplotlib.pyplot as plt
import numpy as np

# Définition de la fonction f(x) = x
def f(x):
    return x

def tracage_corr(x,y,z, matiere):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    # Création d'un tableau numpy pour les valeurs de x
    s= np.linspace(min(x),  max(y), 100)

    # Calcul des valeurs correspondantes de y
    m = f(s)
    
    plt.subplot(1,2,1)
    # Tracé de la fonction 
    plt.plot(s, m, color ='r',label="f(x) = x")

    # Configuration des labels des axes et du titre
    plt.xlabel(f"note en {matiere}")
    plt.ylabel("la note globale")
    plt.title(f"Corrélation entre les notes de {matiere} et la moyenne globale", fontsize=8)

  

    # Tracé d'un nuage de points pour visualiser la corrélation
    plt.plot(x, y, 'b.', label=f"Notes de {matiere}")
    

    # Configuration des limites des axes
    plt.xlim(min(x), max(x))
    plt.ylim(min(y), max(y))

    # Configuration de la légende
    plt.legend()


    plt.subplot(1,2,2)
    
    
    # Configuration des labels des axes et du titre
    plt.xlabel(f"note en {matiere}")
    plt.ylabel("le rang de classement")
    plt.title(f"Corrélation entre la note de {matiere} et le rang de classement", fontsize = 8)
    # Tracé d'un nuage de points pour visualiser la corrélation
    plt.plot(x, z,'b.', label=f'note de {matiere}')
    # Configuration des limites des axes
    plt.xlim(min(x), max(x))
    plt.ylim(min(z), max(z))
    #afficher 
    plt.show()
    #calculer la corrélation
    corre = x.corr(y)   
    corre1 = x.corr(z) 
    print(f"Coefficient de corrélation entre les notes de {matiere} et la moyenne globale:", corre)
    print(f"Coefficient de corrélation entre les notes de {matiere} et le rang de classement:", corre1)
    #Il existe une corrélation entre la moyenne de l'élève dans une matière et son rang dans le classement. En général, les élèves qui ont des notes plus élevées dans ces matières ont tendance à être classer plus haut dans le classement.
```
